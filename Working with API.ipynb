```python
import requests
import pandas as pd

df = pd.DataFrame()  # Initialize an empty DataFrame to store the movie data

for i in range(1, 501):  # Loop from page 1 to 500
    response = requests.get('https://api.themoviedb.org/3/movie/top_rated?api_key=770e15dd197156c37843f6929842ae65&language=en-US&page={}'.format(i))
    
    # Check if the request was successful
    if response.status_code == 200:
        page_data = pd.DataFrame(response.json()['results'])[['id', 'title', 'release_date', 'overview', 'popularity', 'vote_average', 'vote_count']]
        df = df.append(page_data, ignore_index=True)
    else:
        print("Failed to fetch data from page", i)

# Now df contains the movie data from all pages
```
