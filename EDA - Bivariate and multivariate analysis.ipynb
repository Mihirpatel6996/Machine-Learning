# 1. Scatterplot  (Numerical - Numerical) - Bivariate 

```python
sns.scatterplot(data = tips , x='total_bill', y= 'tip' )
```

# 1.a Scatter plot (3 or more numerical or categorical) - Multivariate 


```python
tips.head()
```

```python
sns.scatterplot(data = tips , x='total_bill', y= 'tip', hue= 'sex')
```

```python
sns.scatterplot(data = tips , x='total_bill', y= 'tip',hue='sex', style = 'smoker' )
```

```python
sns.scatterplot(data = tips , x='total_bill', y= 'tip',hue='sex', style = 'smoker', size='size' )
```

by using scatter plot we were able to plot 5 features in one grap i.e  <br>
1) total bill vs tip <br>
2) total bill vs tip with sex(male or feamle) <br>
3) total bill vs tip with sex and if that person is smoker or not <br>
4) total bill vs tip with sex and if that person is smoker or not also size of that customer <br>


# 2. Bar Plot (Numerical - Categorical)- Bivariate

```python
titanic.head()
```

```python
sns.barplot(data= titanic, x = 'Pclass' , y='Age')
```

# 2.b  Bar Plot (Numerical - Categorical)- Multivariate

```python
sns.barplot(data= titanic, x = 'Pclass' , y='Age', hue = 'Sex')
```

# 3. Box Plot (Numerical - Categorical) - Bivariate

```python
sns.boxplot(data= titanic, x = 'Sex',y = 'Age')
```

# 3. Box Plot (Numerical - Categorical) -  Multivariate

```python
sns.boxplot(data= titanic, x = 'Sex',y = 'Age', hue = 'Survived')
```

# 4. Distplot (Numerical - Categorical) - Bivariate

```python
sns.distplot(titanic[titanic['Survived']==0]['Age'])
sns.distplot(titanic[titanic['Survived']==1]['Age'])
```

```python
sns.distplot(titanic[titanic['Survived']==0]['Age'],hist=False)
sns.distplot(titanic[titanic['Survived']==1]['Age'],hist=False)
```

# 5. HeatMap (Categorical - Categorical) - bivariate

```python
pd.crosstab(titanic['Pclass'],titanic['Survived'])
```

```python
sns.heatmap(pd.crosstab(titanic['Pclass'],titanic['Survived']))
```

# 6. ClusterMap (Categorical - Categorical)

```python
sns.clustermap(pd.crosstab(titanic['Parch'],titanic['Survived']))
```

# 7. Pairplot
