# 4. OneHotEncoding with Top Categories

```python
counts = df['brand'].value_counts()
```

```python
counts
```

```python
df['brand'].nunique()

```

```python
threshold = 100

repl = counts[counts <= threshold].index
```

```python
repl
```

```python
df['brand'] = df['brand'].replace(repl, 'uncommon')
```
